```python
import os
import numpy as np
from PIL import Image
import time

from os import listdir
from os.path import isfile, join
from mycode import convert_coco2custom
from PIL import ImageFile, Image

ImageFile.LOAD_TRUNCATED_IMAGES = True

from label_ids import label_colormap

from shutil import copyfile
import time


ROOT_DIR = '/Users/sephidator/Desktop/生成结果/refinedcoco'
NEW_ROOT_DIR = '/Users/sephidator/Desktop/生成结果/enhancedcoco'
```

```python
def have_border(image_path):
    def only_one_color(border):
        return len(np.unique(border[0])) == 1
    
    image_pil = Image.open(image_path).convert('RGB')
    image_numpy = np.asarray(image_pil)
    # h, w, c = image_numpy.shape
    left_border = image_numpy[:][0]
    right_border = image_numpy[:][-1]
    top_border = image_numpy[0][:]
    bottom_border = image_numpy[-1][:]
    return only_one_color(left_border) or only_one_color(right_border) \
        or only_one_color(top_border) or only_one_color(bottom_border)
```

```python
def valid(image_path, label_path):
    if have_border(image_path):
        return False
    
    label_pil = Image.open(label_path)
    label_numpy = np.asarray(label_pil)
    unknown_pixel = (label_numpy == 0).sum()
    w, h = label_pil.size
    unknow_rate = unknown_pixel / (w * h)
    
    return unknow_rate <= 0.02
```

```python
def save_files(image_path, label_path, id):
    id = str(id)
    id = '0' * (5 - len(id)) + id
    new_image_path = f'{NEW_ROOT_DIR}/images/{id}.jpg'
    new_label_path = f'{NEW_ROOT_DIR}/labels/{id}.png'
    new_colorlabel_path = f'{NEW_ROOT_DIR}/colorlabels/{id}.png'
    
    copyfile(image_path, new_image_path)
    copyfile(label_path, new_label_path)
    
    label_pil = Image.open(label_path)
    label_numpy = np.asarray(label_pil)
#     label_numpy = np.transpose(label_numpy, (1, 2, 0))
#     w, h, c = label_numpy.shape
#     label_numpy = label_numpy.reshape((w, h))
    label_numpy = label_colormap[label_numpy].astype(np.uint8)
    colorlabel = Image.fromarray(label_numpy).convert('RGB')
    colorlabel.save(new_colorlabel_path)
```

```python
image_names = sorted(os.listdir(f'{ROOT_DIR}/images'))
label_names = sorted(os.listdir(f'{ROOT_DIR}/labels'))
id = 0

for image_name, label_name in zip(image_names, label_names):
    a = time.time()
    image_path = f'{ROOT_DIR}/images/{image_name}'
    label_path = f'{ROOT_DIR}/labels/{label_name}'
    if valid(image_path, label_path):
        save_files(image_path, label_path, id)
        id += 1
        print(f'{label_name} SAVED', end=', ')
    else:
        print(f'{label_name} IGNORED', end=', ')
        
    print(str(time.time() - a)[:4] + ' sec')
```
